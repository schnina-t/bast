# Tagesganglinien auf Autobahnen und Bundesstraßen für HiMeMo

```python
import geopandas as gpd

import pandas as pd
import altair as alt

from datetime import timedelta
```

read input bast files:

```python
file_A = '2018_A_S.txt'
file_b = '2018_B_S.txt'
```

```python
a_data= pd.read_csv(file_A, sep=';', dtype={'Strnum': 'str'})
#b_data= pd.read_csv(file_B, sep=';', dtype={'Strnum': 'str'})
```

```python
a_data.columns
```

```python
a_data['pt'] = a_data['Pkw_R1']+ a_data['Pkw_R2']
a_data['ft'] = a_data['Lkw_R1']+ a_data['Lkw_R2']
```

Stunde und Datum korrigieren für Eintragungen 24h:

```python
a_data['Stunde']
```

```python
a_data['Stunde'] = a_data.apply(lambda x: 0 if x['Stunde'] == 24 else x['Stunde'], axis=1)
a_data['datetime'] = pd.to_datetime(( ("20" + a_data['Datum'].astype(str)) + '.'+ ("0" + a_data['Stunde'].astype(str)).str[-2:]  ), format='%Y%m%d.%H')
```

```python
a_data['datetime'] = a_data.apply(lambda x: x['datetime'] + timedelta(days=1) if x['Stunde'] == 0 else x['datetime'], axis=1)
a_data[:24]
```

```python
a_data['weekday'] = a_data['datetime'].dt.day_name()
a_data['weekday_no']=a_data['datetime'].dt.weekday
```

```python
a_data = a_data.set_index('datetime')
```

```python
a_data[:24]
```

```python
a_data = a_data[['Strklas','weekday','weekday_no','pt','ft','Stunde']].copy()
a_data.head()
```

```python
a_pt_group = a_data.groupby(['weekday','weekday_no','Stunde'])['pt'].mean()
a_ft_group = a_data.groupby(['weekday','weekday_no','Stunde'])['ft'].mean()

a_pt_group=a_pt_group.reset_index().copy()
a_ft_group=a_ft_group.reset_index().copy()

a_pt_group.rename(columns = {'pt':'load'}, inplace = True)
a_ft_group.rename(columns = {'ft':'load'}, inplace = True)

a_pt_group['type'] = 'pt'
a_ft_group['type'] = 'ft'

a_pt_group['load_rel']=a_pt_group['load']/a_pt_group.groupby(['weekday'])['load'].transform('sum')
a_ft_group['load_rel']=a_ft_group['load']/a_ft_group.groupby(['weekday'])['load'].transform('sum')

a_data_group = a_pt_group.append(a_ft_group, ignore_index=True)

a_data_group.head()
```

```python
sum(a_data_group[a_data_group['weekday_no'] == 0]['load_rel'])
```

```python
a_data_group['x_axis'] = a_data_group['weekday_no']*24 + a_data_group['Stunde']
a_data_group.head()
```

```python
a_data_group.to_csv('a_rel_grouped.csv')
```

Plot:

```python
source=a_data_group

selection1 = alt.selection_multi(fields=['type'])
selection2 = alt.selection_multi(fields=['weekday'])
#selection3 = alt.selection_multi(fields=['Strklas'])#, init={'weekday': 'Monday'})

chart= alt.Chart(source).mark_line(point=True).encode(
    x='Stunde:O',
    y=alt.Y('load_rel:Q', title='share of daily traffic'),
    #color='weekday',
    color=alt.Color('weekday:N', legend=None),
    detail=['type'],
    shape=alt.Shape('type:O', legend=None),
    
    tooltip=[
        #alt.Tooltip('count(Land):Q', title='Messstation mit DTV'),
        #alt.Tooltip('freq', title='Anzahl Zählorte Bundesland'),
        alt.Tooltip('load_rel', title='Anteil an Tagesverkehr')
        ],
    #row='year'
).transform_filter(
    selection1
).transform_filter(
    selection2
#).transform_filter(
#    selection3
).interactive()

## interactive legend
legend_type = alt.Chart(source).mark_point().encode(
    y=alt.Y('type:N', axis=alt.Axis(orient='right', titleX=65.), title='type'),
    color=alt.condition(selection1,
                    alt.value('black'),
                    alt.value('lightgray')),
    shape = alt.Shape('type:O', legend = None),
).add_selection(
    selection1
)

## interactive legend
legend_weekday = alt.Chart(source).mark_point().encode(
    y=alt.Y('weekday:N', title='Weekday', axis=alt.Axis(orient='right'), sort=["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]),
    #color=alt.Color('weekday_name:N', legend=None),
    color=alt.condition(selection2,
                    alt.Color('weekday:N', legend=None),
                    alt.value('lightgray')),
).add_selection(
    selection2
)

## interactive legend
#legend_str = alt.Chart(source).mark_point().encode(
#    y=alt.Y('Strklas:N', axis=alt.Axis(orient='right', titleX=65.), title='Straßenklasse'),
#    #color=alt.Color('weekday_name:N', legend=None),
#    color=alt.condition(selection3,
#                    alt.value('black'),
#                    alt.value('lightgray')),
#        shape=alt.Shape('Strklas:O', legend=None),
#).add_selection(
#    selection3
#)
#


full_ch=(chart | (legend_weekday&legend_type))
```

```python
# adding Titel and Subtitle
title = alt.Chart(
    {"values": [{"text": "Relative Tagesganglinien aus Bast-Zählstellen für Pkw in Deutschland, 2018"}]}
).mark_text(size=20, align='left',strokeWidth=0,stroke='white').encode(  #,anchor='start' , align='left'
    text="text:N",
    x=alt.value(0)
)

subtitle = alt.Chart(
    {"values": [{"text": 'Anteil an Tagesverkehr auf Autobahnen'}]}
).mark_text(size=11, align='left').encode(
    text="text:N",
    x=alt.value(0),
)

quelle = alt.Chart(
    {"values": [{"text": 'Quelle: https://www.bast.de/BASt_2017/DE/Verkehrstechnik/Fachthemen/v2-verkehrszaehlung/Stundenwerte.html'}]}
).mark_text(size=11, align='left').encode(
    text="text:N",
    x=alt.value(330),
    y=alt.value(20)
)
  
fullchart_title= alt.vconcat(
    title,
    subtitle,
    full_ch,
    quelle
).configure_view(
    stroke=None
).configure_concat(
    spacing=10
)

fullchart_title
```

```python
a_data_group['x_axis2'] = a_data_group['weekday'] + '-' + a_data_group['Stunde'].astype('str') + 'h'
```

```python
a_data_group = a_data_group.set_index('x_axis')
a_data_group = a_data_group.sort_index()
```

```python
week = a_data_group['x_axis2'].to_list()
```
